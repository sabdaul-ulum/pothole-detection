```python
import cv2
import os
import glob

# ==========================================
# KONFIGURASI FOLDER
# ==========================================
IMAGE_DIR = '../data/n-rdd2024/test/images'         # Folder sumber gambar
LABEL_DIR = '../data/n-rdd2024/test/labels'         # Folder sumber label txt
OUTPUT_DIR = '../dataset/nrdd2024/visualized_results'  # Folder tujuan untuk menyimpan hasil
CLASS_NAMES = ['Pothole']                  # Nama kelas

# Buat folder output jika belum ada
os.makedirs(OUTPUT_DIR, exist_ok=True)

def process_and_save_yolo(img_path, label_path, output_path):
    # Membaca gambar
    img = cv2.imread(img_path)
    if img is None:
        print(f"[!] Gagal membaca gambar (mungkin rusak/bukan gambar): {img_path}")
        return
        
    img_height, img_width, _ = img.shape

    # Mengecek apakah file label (.txt) ada
    if os.path.exists(label_path):
        with open(label_path, 'r') as f:
            lines = f.readlines()
            
        for line in lines:
            parts = line.strip().split()
            if len(parts) != 5:
                continue
                
            class_id = int(parts[0])
            x_center = float(parts[1])
            y_center = float(parts[2])
            w = float(parts[3])
            h = float(parts[4])

            # Konversi koordinat YOLO ke Piksel
            x_min = int((x_center - w / 2) * img_width)
            y_min = int((y_center - h / 2) * img_height)
            x_max = int((x_center + w / 2) * img_width)
            y_max = int((y_center + h / 2) * img_height)

            # Menggambar Bounding Box 
            # Note: OpenCV menggunakan format warna BGR, bukan RGB. (0, 0, 255) = Merah
            cv2.rectangle(img, (x_min, y_min), (x_max, y_max), (0, 0, 255), 3)
            
            # Menambahkan Teks Label
            label_text = CLASS_NAMES[class_id] if class_id < len(CLASS_NAMES) else str(class_id)
            cv2.putText(img, label_text, (x_min, y_min - 10), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)
    else:
        # Menandai gambar Background (Negative Sample)
        cv2.putText(img, "BACKGROUND (No Pothole)", (20, 40), 
                    cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 0), 3)

    # MENYIMPAN GAMBAR KE FOLDER OUTPUT
    cv2.imwrite(output_path, img)

# ==========================================
# PROSES SEMUA GAMBAR (BATCH PROCESSING)
# ==========================================
print("[*] Mencari data gambar...")

# Mengambil semua file berformat .jpg di dalam folder gambar
# (Ubah '*.jpg' menjadi '*.png' jika format gambarmu png)
image_files = glob.glob(os.path.join(IMAGE_DIR, '*.jpg'))
total_images = len(image_files)

print(f"[*] Menemukan {total_images} gambar. Mulai memproses dan menyimpan...\n")

# Looping untuk mengeksekusi semua gambar
for index, img_file in enumerate(image_files, start=1):
    base_name = os.path.splitext(os.path.basename(img_file))[0]
    txt_file = os.path.join(LABEL_DIR, f"{base_name}.txt")
    
    # Nama file output (misal: image01_cek.jpg)
    out_file = os.path.join(OUTPUT_DIR, f"{base_name}_cek.jpg")
    
    process_and_save_yolo(img_file, txt_file, out_file)
    
    # Menampilkan progress bar sederhana di terminal
    if index % 50 == 0 or index == total_images:
        print(f"    -> Progress: {index} / {total_images} selesai...")

print(f"\n[*] SUKSES! Semua hasil telah disimpan di dalam folder: {OUTPUT_DIR}")
```
